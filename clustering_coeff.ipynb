```python
import networkx as nx 
  
G = nx.Graph()
#No need to add two edges for undirected edges
G.add_edges_from([('A', 'B'), ('A', 'E'), ('A', 'F'),
                  ('B', 'F'), ('B', 'C'), 
                  ('C', 'G'), ('C', 'H'),
                 ('D', 'H'),
                 ('E', 'F'), ('E', 'I'), ('E', 'J'), ('E', 'K'),
                  ('F', 'K'),
                  ('G', 'K'), ('G', 'L'), ('G', 'H'),
                  #('H', 'K'), ('H', 'L'),
                  ('I', 'J'), ('J', 'K')
                  #('K', 'L')
                 ])
  
# returns a Dictionary with clustering value of each node
#print(nx.clustering(G))
nx.draw_networkx(G)
# This returns clustering value of specified node
print("clustering value of specified node:",nx.clustering(G, 'F'))
print("Global Clustering Coefficient:",nx.transitivity(G))
```
