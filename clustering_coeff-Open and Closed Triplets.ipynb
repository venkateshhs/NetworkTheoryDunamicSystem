```python
import networkx as nx 
import networkx.algorithms.triads as np
  
G = nx.Graph()
#No need to add two edges for undirected edges
G.add_edges_from([('A', 'B'), ('A', 'E'), ('A', 'F'),
                  ('B', 'C'),('B', 'F'),
                  ('C', 'G'),('C', 'H'),('C', 'D'),('C', 'F'),
                 ('D', 'H'),
                 ('E', 'F'),('E', 'I'),('E', 'J'),
                 ('F', 'J'),
                 ('G', 'K'),('G', 'L'),
                 ('I', 'J'),
                  ('H', 'K'),
                 ('H', 'L'),
                 ('K', 'L')])
  
# returns a Dictionary with clustering value of each node
#print(nx.clustering(G))
nx.draw_networkx(G)

# This returns clustering value of specified node
print("Local Clustering of a vertex :" , nx.clustering(G, 'F'))

###########################

closed_triplets = nx.triangles(G)
for i in closed_triplets:
    closed_triplets[i] = 2*closed_triplets[i]

print("Closed Triplets : ", closed_triplets)

values = closed_triplets.values()
nCT = sum(values)

print("No. of Closed Triplets : ", nCT)

print(nx.transitivity(G))

CC = nx.transitivity(G)

print("No. of Open Triplets : ", (nCT-(CC*nCT))/CC)
```
