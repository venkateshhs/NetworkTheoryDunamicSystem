```python
import networkx as nx 
  
G = nx.Graph()
 #In Undirected edge graph ,write one edge only once  
G.add_edges_from([('a', 'b'),('a', 'e'),
                  ('b', 'f'),
                  ('c', 'd'), ('c', 'g'),   
                  ('d', 'h'),  ('f', 'g'),
                  ('e', 'f'),('e', 'i'),
                  ('i', 'j'),  ('f', 'k'),  ('k', 'l'),
                 ('g', 'h'),('g', 'j'),
                  ('h', 'l')])

  #[('1', '5'), ('1', '2'),
   #               ('2', '3'), 
   #               ('5', '3'),  ('2', '4'),  ('3', '4'),
    #              ('4', '7'),
    #              ('7', '6')]
              
nx.draw_networkx(G)

print("Shortest Path: ",nx.shortest_path(G,'a','g'))
  
print("Shortest Path Length: ",nx.shortest_path_length(G, 'a', 'g'))

print("Diameter: ",nx.diameter(G))
#Out: 6
#Raius is smallest eccentricity
print("Radius: ",nx.eccentricity(G))
```
